```python
# barh() 세로막대 그래프

# 전체 인구를 막대 그래프로 표현

import matplotlib.pyplot as plt
import csv

file = open('population_enc.csv', 'r', encoding='utf-8')
data = csv.reader(file)

populationList = []
maleList = []
femaleList = []

addr = input('검색하고 싶은 지역의 읍면동 이름을 입력하세요 >>>')


for row in data :
    if addr in row[0] : # 구로구 행정 구역의 인구만 보겠다
        for i in row[3:104] : # 남여 인구 합산 0세부터 100세의 위치 (내가 보려는 실질적인 데이터가 있는곳)
            populationList.append(int(i.replace(',',''))) # 콤마를 제외하고 정수형 데이터로 저장
        for i in row[106:207] : # 남자 0세 ~ 100세
             maleList.append(int(i.replace(',',''))*-1) # 반대방향으로 나오게 하기 위해 -1 을 곱함
        for i in row[209:310] : # 여자 0세 ~ 100세
             femaleList.append(int(i.replace(',','')))
            
print(populationList)
print(maleList)
print(femaleList)

plt.rc('font', family='d2coding')
plt.title('연령별, 성별 인구분포 항아리 그래프')

plt.rcParams['axes.unicode_minus'] = False

plt.barh(range(101), maleList,  label='남성')
plt.barh(range(101), femaleList, label='여성')

#plt.plot(populationList, color = 'gray', label='전체')
#plt.plot(maleList, color = 'blue', label='남자')
#plt.plot(femaleList, color = 'red', label='여자')

plt.legend()

#plt.show()

plt.savefig('populationChart.png', dpi=500) # 차트 저장
file.close()            
```
