```python
import re #문자열 처리 패키지 ( 정규식 패턴 )
import konlpy # 한글 형태소 분석 패키지
import pandas as pd
from wordcloud import WordCloud # 워드 클라우드
import matplotlib.pyplot as plt
import PIL
import numpy as np



hannanum = konlpy.tag.Hannanum()

presidentSpeech = open('presidentSpeetch.txt', 'r', encoding='utf-8').read()
presidentSpeech = re.sub( '[^가-힣]' , ' ', presidentSpeech) # re에서는 sub가 replace 함수와 같다 # 가-힣 : 완성형 글자에서 첫자와 마지막 글자
# 완성형을 제외한 모든 문자를 제거하여 다시 할당해준다
# print(presidentSpeech)

hanNouns = hannanum.nouns(presidentSpeech )
# print(hanNouns)

df = pd.DataFrame({'word' : hanNouns})
df['count'] = df['word'].str.len()
df =df.query('count >= 2') # 단어의 길이(count)가 2 이상 나온 단어에만 애해서
df.sort_values('count')# count를 기준으로 정렬

# 단어별 분리 #  word를 그룹화 한뒤 key 값으로 만들어 중복되는 횟수를 구함. 내림차순 정렬
df = df.groupby('word', as_index=False).agg(n=('word', 'count')).sort_values('n', ascending=False) 
#print(df)

keywordTop20 = df.head(20)
#print(keywordTop20)

dicWord = df.set_index('word').to_dict()['n'] # word를 key로 , n을 value로 하는 딕셔너리로 변환
print(dicWord)

font = 'D2CodingBold.ttf' # 한글 폰트지정 같은 경로에 있어야 한다

#cloud.png 파일로 mask 이미지 만들기
icon = PIL.Image.open('mask3.png')
img = PIL.Image.new('RGB', icon.size, (255,255,255)) # cloud.png 크기의 흰색배경의 새로운 이미지 생성
img.paste(icon, icon) # x , y 좌표에 같은 미지를 붙여넣는다
img = np.array(img)

# 워드 클라우드 객체 생성
wc = WordCloud(random_state = 3000, font_path=font, width=400, height=400, background_color='white', mask=img, colormap ='inferno')

imgWC = wc.generate_from_frequencies(dicWord)

plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(imgWC)
```
