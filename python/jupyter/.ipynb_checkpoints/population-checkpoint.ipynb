```python
import matplotlib.pyplot as plt
import csv

file = open('population_enc.csv', 'r', encoding='utf-8')
data = csv.reader(file)

populationList = []
maleList = []
femaleList = []


for row in data :
    if '1153000000' in row[0] : # 구로구 행정 구역의 인구만 보겠다
        for i in row[3:104] : # 남여 인구 합산 0세부터 100세의 위치 (내가 보려는 실질적인 데이터가 있는곳)
            populationList.append(int(i.replace(',',''))) # 콤마를 제외하고 정수형 데이터로 저장
        for i in row[106:207] : # 남자 0세 ~ 100세
             maleList.append(int(i.replace(',','')))
        for i in row[209:310] : # 여자 0세 ~ 100세
             femaleList.append(int(i.replace(',','')))
            
print(populationList)
print(maleList)
print(femaleList)

plt.rc('font', family='d2coding')
plt.title('연령별, 성별 인구분포 그래프')

plt.plot(populationList, color = 'gray', label='전체')
plt.plot(maleList, color = 'blue', label='남자')
plt.plot(femaleList, color = 'red', label='여자')

plt.legend()

plt.show()
file.close()            
```

```python
# barh() 세로막대 그래프

# 전체 인구를 막대 그래프로 표현

import matplotlib.pyplot as plt
import csv

file = open('population_enc.csv', 'r', encoding='utf-8')
data = csv.reader(file)

populationList = []
maleList = []
femaleList = []

addr = input('검색하고 싶은 지역의 읍면동 이름을 입력하세요 >>>')


for row in data :
    if addr in row[0] : # 구로구 행정 구역의 인구만 보겠다
        for i in row[3:104] : # 남여 인구 합산 0세부터 100세의 위치 (내가 보려는 실질적인 데이터가 있는곳)
            populationList.append(int(i.replace(',',''))) # 콤마를 제외하고 정수형 데이터로 저장
        for i in row[106:207] : # 남자 0세 ~ 100세
             maleList.append(int(i.replace(',',''))*-1) # 반대방향으로 나오게 하기 위해 -1 을 곱함
        for i in row[209:310] : # 여자 0세 ~ 100세
             femaleList.append(int(i.replace(',','')))
            
print(populationList)
print(maleList)
print(femaleList)

plt.rc('font', family='d2coding')
plt.title('연령별, 성별 인구분포 항아리 그래프')

plt.rcParams['axes.unicode_minus'] = False

plt.barh(range(101), maleList,  label='남성')
plt.barh(range(101), femaleList, label='여성')

#plt.plot(populationList, color = 'gray', label='전체')
#plt.plot(maleList, color = 'blue', label='남자')
#plt.plot(femaleList, color = 'red', label='여자')

plt.legend()

#plt.show()

plt.savefig('populationChart.png', dpi=500) # 차트 저장
file.close()            
```

```python
# barh() 세로막대 그래프

# 전체 인구를 막대 그래프로 표현

import matplotlib.pyplot as plt
import csv

file = open('population_city_enc.csv', 'r', encoding='utf-8')
data = csv.reader(file)

populationList = []
maleList = []
femaleList = []


city = []


for row in data :
    city = row[0].split(" ")[0]
    #city = row[0]
    if city in row[0] : # 해당 도시의 인구만 보겠다
        for i in row[3:24] : # 남자 0세 ~ 100세
             maleList.append(int(i.replace(',',''))*-1) # 반대방향으로 나오게 하기 위해 -1 을 곱함
        for i in row[26:47] : # 여자 0세 ~ 100세
             femaleList.append(int(i.replace(',','')))
            
    #print(populationList)
        print(maleList)
        print(femaleList)

        plt.rc('font', family='d2coding')
        plt.title(city +' 연령별, 성별 인구분포 항아리 그래프')

        plt.rcParams['axes.unicode_minus'] = False

        plt.barh(range(21), maleList,  label='남성')
        plt.barh(range(21), femaleList, label='여성')

        #plt.plot(populationList, color = 'gray', label='전체')
        #plt.plot(maleList, color = 'blue', label='남자')
        #plt.plot(femaleList, color = 'red', label='여자')

        plt.legend()

        #plt.show()
        
        plt.savefig( city +'populationChart.png', dpi=500) # 차트 저장
        
        maleList = []
        femaleList = []
        plt.close()

        
file.close()            
```

```python
# 파이(pie) 차트

import matplotlib.pyplot as plt

plt.rc('font', family='d2coding')
count = [3, 0, 1, 3] # A, B, AB, O 형
label= ['A형', 'B형', 'AB형', 'O형' ]
color = ['pink', 'orange', 'skyblue', 'yellow']

plt.pie(count, labels=label, autopct='%.1f%%', explode=(0, 0, 2, 0), colors=color)

plt.legend()
plt.show()
```

```python
# 제주도의 성별 비율을 pie 차트로 그리자

import matplotlib.pyplot as plt
import csv

file = open('population_enc.csv', 'r', encoding='utf-8')
data = csv.reader(file)

populationList = []

#addr = input('검색하고 싶은 지역의 읍면동 이름을 입력하세요 >>>')


for row in data :
    if '5000000000' in row[0] : 
        # row[105] 값을 인구 리스트의 0 번째에 할당
        populationList.insert(0, int(row[105].replace(',','')) )
        # row[208] 값을 인구 리스트의 0 번째에 할당
        populationList.insert(1, int(row[208].replace(',','')) )
            
print(populationList)


plt.rc('font', family='d2coding')
plt.title('제주도 성비 분포도')
label= ['남성', '여성']
color = ['skyblue' , 'pink']

plt.pie(populationList, labels=label, colors=color, autopct = '%.1f%%')

plt.legend()

plt.show()

#plt.savefig('populationChart.png', dpi=500) # 차트 저장
file.close()            
```
