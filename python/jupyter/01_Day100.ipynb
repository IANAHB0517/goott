```python
import csv

# CSV 파일을 읽기 모드로 open 할 수 있는 스트리밍 객체 얻기
file = open('C:\\Users\\tachyon\\Downloads\\seoulTemperature.csv' , 'r' , encoding='utf-8')
data = csv.reader(file, delimiter=',') # ,로 구분하여 스트리밍 객체에서 파일 읽기

# ['날씨', '지정', '평균', '최저', '최고']

max_temp = -99999
max_date = ''

min_temp = 99999
min_date = ''

for row in data :
    if row[-1] == '' : #6.25 기간 중  기온 측정 데이터가 없는 데도 최고 기온을 얻을 수 있도록 조작
        row[-1] = -9999
    row[-1] = float(row[-1])
   
    
    if max_temp < row[-1] : # row의 최고 기온이 현재 가지고 있는 최고 기온 보다 높다면
        max_date = row[0] # row의 0 번째 에서 날짜를 가져와 date에 할당
        max_temp = row[-1] # row의 -1 번째 에서 최고 기온을 가져와 temp에 할당
    
    
    if row[-2] == '' :
        row[-2] = 9999
    row[-2] = float(row[-2])
   
        
    if min_temp > row[-2] :
        min_date = row[0]
        min_temp = row[-2]
        
    print(row)
    
print('가장 더웠던 날 최고기온 : ' , max_date, max_temp)
print('가장 추웠던 날 최저기온 : ' , min_date, min_temp)

file.close()
```

```python
# 오늘 날짜 (5월, 8일)의 기온 변화(최고기온)를 그래프로 그려보자

import csv
import matplotlib.pyplot as plt

# CSV 파일을 읽기 모드로 open 할 수 있는 스트리밍 객체 얻기
file = open('C:\\Users\\tachyon\\Downloads\\seoulTemperature.csv' , 'r' , encoding='utf-8')
data = csv.reader(file, delimiter=',') # ,로 구분하여 스트리밍 객체에서 파일 읽기

# ['날씨', '지정', '평균', '최저', '최고']

maxTempList = []
minTempList = []

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '05' and row[0].split('-')[2]=='08':
            maxTempList.append(float(row[-1]))
            minTempList.append(float(row[-2]))
        
print(maxTempList)
print(len(maxTempList))

plt.rc('font', family='d2coding') # title을 한글로 적기 위해 폰트 지정 !! 꼭 사용할 폰트를 명시해주어야 한다
plt.title('5월8일 기온변화 그래프')
plt.plot(maxTempList, 'r', label = 'maxTemp')
plt.plot(minTempList, 'b', label = 'minTemp')
plt.legend()
plt.show()

file.close()
```

```python
# 1월과 8월의 최고 기온을 각각 히스트 그램으로 그려보자


import csv
import matplotlib.pyplot as plt

# CSV 파일을 읽기 모드로 open 할 수 있는 스트리밍 객체 얻기
file = open('C:\\Users\\tachyon\\Downloads\\seoulTemperature.csv' , 'r' , encoding='utf-8')
data = csv.reader(file, delimiter=',') # ,로 구분하여 스트리밍 객체에서 파일 읽기

# ['날씨', '지정', '평균', '최저', '최고']

january = []
august = []

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '01' :
            january.append(float(row[-1]))
        if row[0].split('-')[1] == '08' :
            august.append(float(row[-1]))
        



plt.rc('font', family='d2coding') # title을 한글로 적기 위해 폰트 지정 !! 꼭 사용할 폰트를 명시해주어야 한다
plt.title(' 1월 8월의 최고기온 빈도수 그래프')
plt.hist(august, color= 'r', bins=100 , label='August')

plt.hist(january, color= 'b', bins=100, label='January')
plt.legend()
plt.show()

file.close()
```
