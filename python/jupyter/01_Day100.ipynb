```python
import csv

# CSV    open      
file = open('C:\\Users\\tachyon\\Downloads\\seoulTemperature.csv' , 'r' , encoding='utf-8')
data = csv.reader(file, delimiter=',') # ,     

# ['', '', '', '', '']

max_temp = -99999
max_date = ''

min_temp = 99999
min_date = ''

for row in data :
    if row[-1] == '' : #6.25              
        row[-1] = -9999
    row[-1] = float(row[-1])
   
    
    if max_temp < row[-1] : # row         
        max_date = row[0] # row 0     date 
        max_temp = row[-1] # row -1      temp 
    
    
    if row[-2] == '' :
        row[-2] = 9999
    row[-2] = float(row[-2])
   
        
    if min_temp > row[-2] :
        min_date = row[0]
        min_temp = row[-2]
        
    print(row)
    
print('    : ' , max_date, max_temp)
print('    : ' , min_date, min_temp)

file.close()
```

```python
#   (5, 8)  ()  

import csv
import matplotlib.pyplot as plt

# CSV    open      
file = open('C:\\Users\\tachyon\\Downloads\\seoulTemperature.csv' , 'r' , encoding='utf-8')
data = csv.reader(file, delimiter=',') # ,     

# ['', '', '', '', '']

maxTempList = []
minTempList = []

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '05' and row[0].split('-')[2]=='08':
            maxTempList.append(float(row[-1]))
            minTempList.append(float(row[-2]))
        
print(maxTempList)
print(len(maxTempList))

plt.rc('font', family='d2coding') # title      !!     
plt.title('58  ')
plt.plot(maxTempList, 'r', label = 'maxTemp')
plt.plot(minTempList, 'b', label = 'minTemp')
plt.legend()
plt.show()

file.close()
```

```python
# 1 8      


import csv
import matplotlib.pyplot as plt

# CSV    open      
file = open('C:\\Users\\tachyon\\Downloads\\seoulTemperature.csv' , 'r' , encoding='utf-8')
data = csv.reader(file, delimiter=',') # ,     

# ['', '', '', '', '']

january = []
august = []

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '01' :
            january.append(float(row[-1]))
        if row[0].split('-')[1] == '08' :
            august.append(float(row[-1]))
        



plt.rc('font', family='d2coding') # title      !!     
plt.title(' 1 8   ')
plt.hist(august, color= 'r', bins=100 , label='August')

plt.hist(january, color= 'b', bins=100, label='January')
plt.legend()
plt.show()

file.close()
```
